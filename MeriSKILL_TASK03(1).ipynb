### <div style="text-align: center; background-color:skyblue ; font-family:Impact; color: purple; padding: 20px; line-height: 1;border-radius:10px"> Import Libraries üìö</div>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## <div style="text-align: center; background-color:skyblue ; font-family:Impact; color: purple; padding: 20px; line-height: 1;border-radius:10px">Loading data Set and Preprocessing Of Data(Data Cleaning)</div>

```python
HR_data=pd.read_csv("C:\\Users\\Ujjaval Raj\\Downloads\\HR-Employee-Attrition.csv")
HR_data_head=HR_data.head()
HR_data_tail=HR_data.tail()
print(HR_data_head)
print(HR_data_tail)
```

#### <font color='blue'>Checking for null values and duplicates</font>

```python
HR_data_null=HR_data.isnull().sum()
HR_data_duplicate=HR_data.duplicated().sum()
print(HR_data_null)
print(HR_data_duplicate)
```

#### <font color='blue'>Obtaining Datatypes and Describing data set with basic statistics</font>

```python
HR_data.info()
HR_data.describe()
```

#### <font color='blue'>Deleting column standardHours and Over18 as data is repeated for all rows(relundancy) .</font>

```python
HR_data = HR_data.drop(columns=['Over18','StandardHours'])
HR_data.head(1)
```

#### <font color='blue'>Renaming the columns.</font>

```python
HR_data = HR_data.rename(columns={'NumCompaniesWorked': 'NumberOfWorkedCompanies'})
HR_data.head(1)
```

#### <font color='blue'>Dropping Not available values (NAN).</font>

```python
HR_data = HR_data.dropna()
```

## <div style="text-align: center; background-color:skyblue ; font-family:Impact; color: purple; padding: 20px; line-height: 1;border-radius:10px">Visualisation of the HR data</div>

### <font color='blue'>Plot a correlation map for all numeric variables</font>

```python
numeric_columns = HR_data.select_dtypes(include=['float64', 'int64']).columns

# Calculate the correlation matrix for numeric columns only
correlation_matrix = HR_data[numeric_columns].corr()

sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 8))
cmap = sns.diverging_palette(240, 10, as_cmap=True)
sns.heatmap(correlation_matrix, annot=True, cmap=cmap, fmt=".2f", linewidths=.5)

plt.title('‚ö° Correlation Map for Numeric Variables ‚ö°', fontsize=16, fontweight='bold')

plt.show()
```

### <font color='blue'>Plot a correlation map for all numeric variables</font>

```python
# Categorical Variables Visualizations
categorical_variables = ['Attrition','OverTime','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus']

sns.set_theme(style="whitegrid")

# Iterate through categorical variables for visualizations
for variable in categorical_variables:
    
    plt.figure(figsize=(10, 6))
    sns.countplot(x=variable, data=HR_data, palette='Set2')
    plt.title(f'üìä Distribution of {variable} üìä', fontsize=14, fontweight='bold')
    
    
    plt.show()

```

### <font color='blue'>Plot a correlation map for all numeric variables</font>

```python
# Visualizations for numeric variables
numeric_variables = ['Age', 'DailyRate','TotalWorkingYears', 'Education', 'EducationField','EmployeeCount', 'DistanceFromHome']

for variable in numeric_variables:
    plt.figure(figsize=(10, 6))
    sns.histplot(HR_data[variable], bins=20, kde=True, color='skyblue')
    plt.title(f'Distribution of {variable}')
    plt.show()
```

### <font color='blue'>Relation between Overtime and Age</font>

```python

plt.figure(figsize=(12, 8))
sns.boxplot(x='OverTime', y='Age', data=HR_data, palette='Set3')
plt.title('Relation between Overtime and Age')
plt.show()
```

### <font color='blue'>Relation between BuisenessTravel and JObSatisfaction</font>

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='BusinessTravel', y='JobSatisfaction', data=HR_data, palette='Blues')
plt.title('Business Travel vs. Job Satisfaction')
plt.show()
```

### <font color='blue'>Relation between MaritalStatus and TotalWorkingYears</font>

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='MaritalStatus', y='TotalWorkingYears', data=HR_data, palette='husl')
plt.title('Marital Status vs. Total Working Years')
plt.show()
```
